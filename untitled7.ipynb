```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report


file_path = '/content/processed_flood_dataset.csv.csv'
df = pd.read_csv(file_path)
print(df.head())

```

```python

print(df.isnull().sum())
df.fillna(method='ffill', inplace=True)
```

```python
features = ['Rain', 'Temp Max', 'Temp Min', 'TempRange', 'Rain_3day', 'Rain_7day']
X = df[features]
flood_threshold = 150
df['Flood'] = (df['Rain_7day'] > flood_threshold).astype(int)
y = df['Flood']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
print(classification_report(y_test, y_pred))
```

```python
import joblib
import pandas as pd
import requests
from IPython.display import Audio, display

model = joblib.load('/content/flood_model.pkl')
def fetch_current_weather(api_key, location):
    url = f"http://api.weatherapi.com/v1/current.json?key={api_key}&q={location}&aqi=no"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print("Error fetching data from WeatherAPI")
        return None

def preprocess_weather_data(weather_data):

    temp_max = weather_data['current']['temp_c']
    temp_min = temp_max - 5
    rain = weather_data['current']['precip_mm']

    temp_range = temp_max - temp_min
    rain_3day = rain
    rain_7day = rain

    df = pd.DataFrame([[rain, temp_max, temp_min, temp_range, rain_3day, rain_7day]],
                      columns=['Rain', 'Temp Max', 'Temp Min', 'TempRange', 'Rain_3day', 'Rain_7day'])

    return df

def predict_flood_risk(df):
    prediction = model.predict(df)
    return prediction[0]

def play_sound_alert(flood_risk):
    if flood_risk == 1:
        print("High flood risk! Playing alert sound.")
        display(Audio('/content/WhatsApp Audio 2024-08-27 at 7.26.17 PM (1).mp3', autoplay=True))
    else:
        print("Low flood risk. Playing alert sound.")
        display(Audio('/content/WhatsApp Audio 2024-08-27 at 7.26.17 PM.mp3', autoplay=True))

api_key = '9c19444d82cc488e99b142625241909'
location = 'Chennai'

weather_data = fetch_current_weather(api_key, location)

if weather_data:

    preprocessed_data = preprocess_weather_data(weather_data)

    flood_risk = predict_flood_risk(preprocessed_data)

    play_sound_alert(flood_risk)
```
